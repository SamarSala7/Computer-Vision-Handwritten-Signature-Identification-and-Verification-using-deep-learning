```python
X_train, y_train = np.stack(df_train['gray_images'].to_numpy()), np.stack(df_train['label'].to_numpy())
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2], 1)
X_test, y_test = np.stack(df_test['gray_images'].to_numpy()), np.stack(df_test['label'].to_numpy())
X_test = X_test.reshape(X_test.shape[0], X_train.shape[1], X_train.shape[2], 1)

# X_train, y_train = shuffle(X_train, y_train, random_state=42)
# X_test, y_test = shuffle(X_test, y_test, random_state=42)
```

```python
def get_uncompiled_model():
    input_shape = X_train.shape[1:]
    model = Sequential([
            Conv2D(32, 3, padding="same", activation='relu', input_shape=input_shape),
            BatchNormalization(),
            MaxPooling2D(2),
            Dropout(0.3),
            Conv2D(64, 3, padding="same", activation='relu'),
            BatchNormalization(),
            MaxPooling2D(2),
            Dropout(0.3),
            Flatten(),
            Dense(64, activation = 'relu'),
            Dropout(0.2),
            Dense(1, activation='sigmoid')
    ])
    return model

def get_compiled_model():
    model = get_uncompiled_model()
    model.compile(optimizer=tf.keras.optimizers.Adam(),
        loss=tf.keras.losses.BinaryCrossentropy(),
        metrics=['accuracy'])
    return model

def train_model(X, y, batch_size=4, epochs=20):
    model = get_compiled_model()
    history = model.fit(X, y,
                        batch_size = batch_size,
                        epochs=epochs, verbose=1 
                    )
    return history, model

def evaluate_model(model, X_test, y_test):
    model.evaluate(X_test, y_test, verbose=1, batch_size=1)
    y_pred = model.predict(X_test, verbose=1, batch_size=1)
    y_pred = y_pred.astype(np.int8)
    return y_pred
    
```

```python
import time
models = {
    'model':[],
    'itteration':[]
}
histories = []
y_preds = []
X = [X_train[0:40], X_train[40:80], X_train[80:120], X_train[120:160], X_train[160:200]]
y = [y_train[0:40], y_train[40:80], y_train[80:120], y_train[120:160], y_train[160:200]]
y_t = [y_train[0:8], y_train[8:16], y_train[16:24], y_train[24:32], y_train[32:40]]
X_t = [X_train[0:8], X_train[8:16], X_train[16:24], X_train[24:32], X_train[32:40]]
for i in range(0, 5):
    start_train = time.time()
    print(f'Itteration {i}')
    print('Training')
    history, model = train_model(X[i], y[i])
    end_train = time.time()
    histories.append(history)
    models['model'].append(model)
    models['itteration'].append(i)
    print('Evaluating')
    start_test = time.time()
    y_pred = evaluate_model(model, X_t[i], y_t[i])
    y_preds.append(y_pred)
    end_test = time.time()
    print(f'Train time = {(end_train-start_train):.2f} seconds, Testing time = {(end_test-start_test):.2f} seconds.')
    print('-'*50)
```
